```python
# Import Dependencies
import pandas as pd
```

```python
bitcoin_csv = "Resources/bitcoin_cash_price.csv"
dash_csv = "Resources/dash_price.csv"
```

```python
bitcoin_df = pd.read_csv(bitcoin_csv)
dash_df = pd.read_csv(dash_csv)
```

```python
bitcoin_df.head()
```

```python
dash_df.head()
```

```python
# Merge the two DataFrames together based on the Dates they share

merge_df = pd.merge(bitcoin_df, dash_df, on="Date", how="inner", suffixes=("_Bitcoin", "_Dash"))
merge_df.head()
```

```python
# Collecting best open for Bitcoin and Dash

best_open_bitcoin = merge_df["Open_Bitcoin"].max()
best_open_dash = merge_df["Open_Dash"].max()

# Collecting best close for Bitcoin and Dash

best_close_bitcoin = merge_df["Close_Bitcoin"].max()
best_close_dash = merge_df["Close_Dash"].max()

# Collecting the total volume for Bitcoin and Dash

bitcoin_volume = round(merge_df["Volume_Bitcoin"].sum()/1000000, 2)
dash_volume = round(merge_df["Volume_Dash"].sum()/1000000, 2)
```

```python
# Creating a summary DataFrame using above values

summary_df = pd.DataFrame({
    "Best_Open_Bitcoin": [best_open_bitcoin], "Best_Open_Dash": [best_open_dash],
    "Best_Close_Bitcoin": [best_close_bitcoin], "Best_Close_Dash": [best_close_dash],
    "Total_Volume_Bitcoin (millions)": [bitcoin_volume], "Total_Volume_Dash (millions)": [dash_volume]
})

summary_df
```
